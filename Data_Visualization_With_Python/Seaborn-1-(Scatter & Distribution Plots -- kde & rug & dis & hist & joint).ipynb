## displot() --> histogram

```python
sns.displot(data=mpg, x="weight");  # histogram
```

```python
sns.histplot(data=mpg, x="weight");  # histogram
```

```python
fig, ax = plt.subplots(1,2, figsize=(8,6))
sns.scatterplot(data=tips, x="total_bill", y="tip",s=200, hue="sex", alpha=0.7, ax=ax[0])
sns.histplot(data=mpg, x="weight", ax=ax[1]);

# sns.displot(data=mpg, x="weight", ax=ax[1]);
# çoklu grafiklerde displot çalışmaz
```

```python
sns.displot(data=mpg, x="weight", kind="hist");

# displot default olarak histplot çizer.
# ama kde vb farklı plotlamalar da yapabilir.
```

```python
sns.displot(data=mpg, x="weight", kind="kde");   # displot ile kde çizdirdik
```

```python
sns.displot(data=mpg, x="weight", kde=True);  # kde yi histogram üzerine çizebiliriz
```

```python
sns.displot(data=mpg, x="weight", kind="hist", aspect=1.5);  # aspect ile boyut ayarlayabiliyoruz
```

```python
sns.displot(data=mpg, x="weight", kde=False, aspect=1.5);
```

```python
sns.displot(data=mpg, x="weight", kind="kde", aspect=1.5);
```

```python
# plt.figure(figsize=(8,6))
sns.histplot(data=mpg, x="weight", bins=30);
```

```python
fig, ax = plt.subplots(1,2)

sns.rugplot(data=mpg, x="weight", ax=ax[0])

sns.histplot(data=mpg, x="weight", ax=ax[1], kde=True)
```

```python
plt.subplot(1,2,1)
sns.rugplot(data=mpg, x="weight")

plt.subplot(1,2,2)
sns.histplot(data=mpg, x="weight", kde=True)
```

```python
sns.set(style="darkgrid")

sns.histplot(data=mpg, x="weight", kde=True)
```
