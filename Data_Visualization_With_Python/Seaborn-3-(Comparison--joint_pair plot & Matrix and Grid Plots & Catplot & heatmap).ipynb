```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import warnings
warnings.filterwarnings('ignore') 

import seaborn as sns
import numpy as np
print(sns.get_dataset_names())
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

## catplot()

```python
sns.catplot(data=penguins, x="species", y="bill_length_mm")
```

```python
sns.catplot(data=penguins, x="species", y="bill_length_mm", kind="bar")
```

```python
sns.catplot(data=penguins, x="species", y="bill_length_mm", kind="bar", estimator = np.count_nonzero)
```

```python
 sns.catplot(data=penguins, x="species", y="bill_length_mm", kind="bar", estimator = np.max)
```

```python
 sns.catplot(data=penguins, x="species", y="bill_length_mm", kind="bar", estimator = np.median, ci="sd")
```

```python
 sns.catplot(data=penguins, x="species", y="bill_length_mm", kind="bar", estimator = np.median, ci="sd", col="island")
```

```python
 sns.catplot(data=penguins, x="species", y="bill_length_mm", kind="bar", estimator = np.median, ci="sd", col="island", hue="sex")
```

```python
 sns.catplot(data=penguins, x="species", y="bill_length_mm", kind="box", ci="sd", col="island", hue="sex")
```

```python
 sns.catplot(data=penguins, x="island", y="bill_length_mm", kind="box", ci="sd", col="species", hue="sex")
```

```python
 sns.catplot(data=penguins, x="island", y="bill_length_mm", kind="box", ci="sd", row="species", hue="sex")
```

## PairGrid()

```python
g = sns.PairGrid(penguins)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(penguins)
g.map_diag(sns.kdeplot)
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map_diag(sns.histplot)
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
sns.set_palette(["green","blue", "red"])

g = sns.PairGrid(penguins, hue="species")
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)

g.add_legend()
```

```python
sns.set_palette(["green","blue", "red"])

g = sns.PairGrid(penguins, hue="species", diag_sharey=False)
g.map_diag(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)

g.add_legend()
```

## FacetGrid

```python
g = sns.FacetGrid(data=penguins, col="species", row="island", hue="sex")
g.map(sns.kdeplot, "bill_length_mm")
```

```python
g = sns.FacetGrid(data=penguins, col="species", row="island", hue="sex")
g.map(sns.scatterplot, "bill_length_mm", "body_mass_g")
```
