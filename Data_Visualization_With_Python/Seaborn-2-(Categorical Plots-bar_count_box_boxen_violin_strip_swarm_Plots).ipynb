## Countplot()

A simple plot, it merely shows the total count of rows per category.

```python
sns.countplot(x="day", data=tips);
```

```python
fig, ax = plt.subplots()

ax = sns.countplot(x="day", data=tips);
# sns.countplot(x="day", data=tips, ax=ax) --> istersek bu şekilde de yazabiliriz.

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+1));
```

```python
fig, ax = plt.subplots(figsize=(16,6))

ax = sns.countplot(x="day", data=tips, hue="smoker");
# sns.countplot(x="day", data=tips, ax=ax) --> istersek bu şekilde de yazabiliriz.

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.17, p.get_height()+1));
```

```python
fig, ax = plt.subplots(figsize=(16,6))

ax = sns.countplot(x="day", data=tips, hue="size");
# sns.countplot(x="day", data=tips, ax=ax) --> istersek bu şekilde de yazabiliriz.

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.04, p.get_height()+1));
```

```python
fig, ax = plt.subplots()

ax = sns.countplot(x="day", data=tips);
# sns.countplot(x="day", data=tips, ax=ax) --> istersek bu şekilde de yazabiliriz.

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+1));
    ax.annotate("%" + str(round(p.get_height()/tips.day.count()*100,2)), (p.get_x()+0.15, p.get_height()-5)); # yüzdelik ekledik
```

## barplot()
So far we've seen the y axis default to a count (similar to a .groupby(x_axis).count() call in pandas). We can expand our visualizations by specifying a specific continuous feature for the y-axis. Keep in mind, you should be careful with these plots, as they may imply a relationship continuity along the y axis where there is none

```python
sns.barplot("sex", "tip", data=tips);
```

```python
sns.barplot("sex", "tip", data=tips, estimator=np.median);  # mean, median, sum....
```

```python
sns.barplot("sex", "tip", data=tips, estimator=np.median, ci="sd", hue="smoker"); 
```

## Barplot vs Countplot

```python
fig, ax = plt.subplots(1,2, figsize=(16,6))

sns.barplot(data=tips,   x="day", y="total_bill", ax=ax[0])
sns.countplot(data=tips, x="day",                 ax=ax[1])
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
sns.barplot(x="day", y="total_bill", data=tips, ax = ax[0], estimator=np.sum)   # estimator param (np.mean, np.median, np.sum, np,maz, np.min, np.count_nonzero)
sns.countplot(x="day", data=tips, ax = ax[1])
plt.tight_layout()
```

As described in the video, a boxplot display distribution through the use of quartiles and an IQR for outliers.

```python
sns.boxplot(x="day", y="tip", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

## Violinplot
A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x='day', y="total_bill",data=tips)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(y='day',x='total_bill',data=tips,hue='sex',split=True)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True, bw=0.2)  #bw = bandwidth - bins
```

## boxenplot (letter-value plot)
Official Paper on this plot: https://vita.had.co.nz/papers/letter-value-plot.html

This style of plot was originally named a “letter value” plot because it shows a large number of quantiles that are defined as “letter values”. It is similar to a box plot in plotting a nonparametric representation of a distribution in which all features correspond to actual observations. By plotting more quantiles, it provides more information about the shape of the distribution, particularly in the tails.

```python
sns.boxenplot(x='day',y='tip',data=tips)
```
